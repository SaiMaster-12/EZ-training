```python
class int:
  def __add__(self, nm1, nm2):
    print("sum:",end=" ")
    return nm1 + nm2
  def __sub__(self, nm1, nm2):
    print("sub:",end=" ")
    return nm1 - nm2
  def __mul__(self, nm1, nm2):
    print("mul:",end=" ")
    return nm1 * nm2
  def __div__(self, nm1, nm2):
    print("div:",end=" ")
    return nm1 / nm2
  def __mod__(self, nm1, nm2):
    print("mod:",end=" ")
    return nm1 % nm2
  def __pow__(self, nm1, nm2):
    print("pow:",end=" ")
    return nm1 ** nm2

int1 = int()
print(int1.__add__(20, 10))
print(int1.__sub__(20, 10))
print(int1.__mul__(20, 10))
print(int1.__div__(20, 10))
print(int1.__mod__(20, 10))
print(int1.__pow__(20, 10))
print(type(int1))
```

```python
class Stud:
  def __init__(self, name, roll, usn, marks):
    self.name = name
    self.roll = roll
    self.usn = usn
    self.marks = marks
  def __str__(self):
    return f"Name: {self.name}\nRoll: {self.roll}\nUSN: {self.usn}\nMarks: {self.marks}"
stu_instance  = Stud('sai',35,35,[100,90,45,88,92])
n = stu_instance.__str__()
print(n)
```
