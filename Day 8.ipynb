```python
import numpy as np
```

```python
# Bubble sort

num = int(input("Enter number of elements: "))
print("Enter elements: ")
# arr = np.array([])
arr = np.zeros(num, dtype=int)
for i in range(num):
  j = int(input())
  arr[i] = j

n = len(arr)
for i in range(n-1):
    for j in range(n-1-i):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
print("\nSorted array: ",arr)
```

```python
# Selection sort

unsort = [8,2,24,16,6]
sort = []
while unsort:
  min_val = min(unsort)
  sort.append(min_val)
  unsort.remove(min_val)
print("Sorted list: ",sort)
```

```python
# Quick sort

# If i found smaller element by pivot increase j by 1 and swap it by i.
# If i found bigger element than pivot i will do nothing.
```

```python
l = [8,6,3,1,2,9,4,5,7]
def divide(l,low,high):
  p = l[high]
  pi = high
  j = low - 1
  for i in range(0,high):
    if l[i] <= p:
      j += 1
      l[i],l[j] = l[j],l[i]
  j += 1
  l[j],l[pi] = l[pi],l[j]
  pi = j
  return pi
```

```python
def quicksort(l,low,high):
  if low < high:
    pi = divide(l,low,high)
    quicksort(l,low,pi-1)
    quicksort(l,pi+1,high)
  return
```
