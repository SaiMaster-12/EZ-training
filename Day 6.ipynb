```python
# Problem 1
def count_vovels(s):
    vov = {
        'a': 0,
        'e': 0,
        'i': 0,
        'o': 0,
        'u': 0
    }
    for i in s:
        if i == 'a' or i == 'A':
            vov['a'] += 1
        elif i == 'e' or i == 'E':
            vov['e'] += 1
        elif i == 'i' or i == 'I':
            vov['i'] += 1
        elif i == 'o' or i == 'O':
            vov['o'] += 1
        elif i == 'u' or i == 'U':
            vov['u'] += 1

    max_vov = max(vov.values())
    result = []
    for i, j in vov.items():
        if j == max_vov:
            result.append(i)
    return result

i_p = [
    ["sai", "hey buddy how are you"],
    ["uday", "good bro how about you and others"],
    ["dinesh", "are you programmer"],
    ["pavan", "ya bro its good to be a programmer"],
    ["abdul", "i have purchased a new car"]
]

for i in i_p:
    name, sentence = i
    result = count_vovels(sentence)
    print(f"[{name}]: {result}")
```
