```python
class stack:
    def __init__(self):
        self.items=[]
    def push(self,data):
        self.items.append(data)
    def pop(self):
        self.items.pop()
    def size(self):
        return len(self.items)
    def top(self):
        return self.items[-1]
    def is_empty(self):
        return len(self.items)==0
s=stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.push(50)
print(s.items)
s.pop()
print(s.items)
s.pop()
print(s.items)
s.pop()
print(s.items)
s.pop()
print(s.items)
s.pop()
print(s.items)
```

```python
class Node:
  def __init__(self,data):
    self.value = data
    self.next = None

Head = Node(10)
Head.next = Node(20)
Head.next.next = Node(30)
Head.next.next.next = Node(40)
Head = Tail = Node(10)
Tail.next = Node(20)
Tail.next.next = Node(30)
Tail.next.next.next = Node(40)

print(Tail)
print(Head)
print(Head.next)
print(Head.next.next)
print(Head.next.next.next)
```

```python
def print_LL(Head):
  if Head == None:
    print("Empty list")
    return
  else:
    curr = Head
    while curr != None:
      print(curr.value)
      curr = curr.next
print_LL(Head)
```

```python
# At Initial state
def Front_Insert(Head,data):
  newnode = Node(data)
  newnode.next = Head
  Head = newnode
  print_LL(Head)
  return Head
# newnode = Node(90)
# curr = Head
# Head = newnode
# Head.next = curr
# print_LL(Head)
Front_Insert(Head,100)
```

```python
# Insertion in Between
def inbetwe(Head,data,pos):
  newnode = Node(data)
  curr = Head
  for i in range(pos-1):
    curr = curr.next
  newnode.next = curr.next
  curr.next = newnode
  print("Items in list: ")
  print_LL(Head)

inbetwe(Head,150,3)
```

```python
# Insertion at last
def atlast(Head,data):
  newnode = Node(data)
  curr = Head
  while curr.next != None:
    curr = curr.next
  curr.next = newnode
  print("Items in list: ")
  print_LL(Head)

atlast(Head,420)
```

```python
# Delete at front
def del_front(Head):
  Head = Head.next
  print("Items in list: ")
  print_LL(Head)
del_front(Head)
```

```python
# Deletion in between
def del_betwe(Head,pos):
  curr = Head
  temp = Head
  for i in range(pos-1):
    curr = curr.next
    temp = temp.next
  temp = temp.next
  curr.next = temp.next
  print("Items in list: ")
  print_LL(Head)

del_betwe(Head,3)
```

```python
# Deletion at last
def del_last(Head):
  curr = Head
  temp = Head
  while curr != None:
    curr = curr.next
    if curr.next == None:
      break
    temp = temp.next
  temp.next = None
  print("Items in list: ")
  print_LL(Head)

del_last(Head)
```
