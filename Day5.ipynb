```python
def check_prime(n):
  for i in range(2,n):
    if n%i == 0:
      return 0
  return 1

result = []
N = int(input("Enter the number:"))
flag = 0
k = N+1
while flag<1:
  flag = check_prime(k)
  if flag == 1:
    result.append(k)
    print(k,"Smallest prime number")
  k += 1
```
